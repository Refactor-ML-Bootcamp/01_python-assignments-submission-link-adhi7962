### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
chipo.head(5)
```

### Step 4. How many products cost more than $10.00?

```python
chipo[chipo["item_price"].apply(lambda x: float(x[1:]) > 10.00)]
#chipo["item_price"].apply(lambda x: float(x[1:]) > 10.00)
#chipo.loc[chipo["item_name"] == "Izze"]
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
price = [float(x[1:]) for x in chipo.item_price]
chipo["price"] = price
chipo[chipo["price"]== 44.25]

#asd.quantity
#max(price)
```
